```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

Створюємо символи для змінної та параметрів

```python
x = sp.symbols('x')
a, b = sp.symbols('a b', real=True, positive=True)
```

Визначаємо логістичну функцію

```python
logistic_function = a / (1 + sp.exp(-b * x))
logistic_function
```

Обчислюємо похідну логістичної функції

```python
derivative = sp.diff(logistic_function, x)
derivative
```

Створюємо функції для обчислення значень логістичної функції та її похідної


```python
logistic = sp.lambdify((x, a, b), logistic_function, 'numpy')
logistic_derivative = sp.lambdify((x, a, b), derivative, 'numpy')
```

Задаємо параметри a і b

```python

a_val = 1
b_val = 1
```

Створюємо масив значень x


```python

x_values = np.linspace(-15, 15, 100)
```

Обчислюємо значення логістичної функції та її похідної


```python

y_logistic = logistic(x_values, a_val, b_val)
y_derivative = logistic_derivative(x_values, a_val, b_val)
```

Будуємо графіки

```python

plt.figure(figsize=(5, 5))

plt.plot(x_values, y_logistic, label='Logistic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()


plt.plot(x_values, 3 * y_derivative, label="Derivative", linestyle='--')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()

plt.show()
```
